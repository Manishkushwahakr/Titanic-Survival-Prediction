```python
#DATA CLEANING
#drop useless columns
df.drop(columns=['PassengerId','Name','Ticket','Cabin'], inplace=True,errors='ignore')

df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Fare'].fillna(df['Fare'].mean(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
#Handled missing value in age and fare using mean. remove unnecessarycolumns
```

```python
df['Sex']=df['Sex'].astype(str).str.lower().str.strip()
df['Sex']=df['Sex'].map({'male':0,'female':1})
if 'Embarked' in df.columns:
    df['Embarked']=df['Embarked'].astype(str).str.lower().str.strip()
    df['Embarked']=df['Embarked'].map({'s':0,'c':1,'q':2})

    df=pd.get_dummies(df,columns=['Embarked'],drop_first=True)
```

```python
#converted categorical data into numerical format
print(df['Sex'].value_counts())
```

```python
#feature engineering
#create new useful feature
df['FamilySize']=df['SibSp']+df['Parch'] + 1
```

```python
#created a new feature 'family size, by combining SibSp  Parch to improve model performance
print(df[['SibSp','Parch','FamilySize']].head())
```

### **Model Building**

```python
x= df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#Accuracy and confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy:", accuracy_score(y_test,y_pred))
print("Confusion matrix :",confusion_matrix(y_test,y_pred))
```
